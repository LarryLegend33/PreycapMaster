```python
from csv_analysis import BayesDB_Simulator, stim_analyzer, stim_conditionals
%matplotlib inline
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
from matplotlib import pyplot as pl
import pandas as pd
import numpy as np
import seaborn as sb
from scipy.stats import norm, ttest_ind, ks_2samp
import warnings
warnings.filterwarnings('ignore')
cpal = sb.color_palette()
```

```python
#fish_id = "091418_bdb"
#file_id = "bdb_stimuli.bdb"
#bdsim = BayesDB_Simulator(fish_id, file_id, -1)
stim_data = pd.read_csv('wik_bdb/all_stimuli.csv')
random_stim_data = pd.read_csv('wik_bdb/all_stimuli_random.csv')
hittypes = [1,2,3]
az_blue = [27/255.0, 117/255.0, 187/255.0]
alt_yellow = [255/255.0, 221/255.0, 21/255.0]
sb.set()

```

```python
varb = "Alt Coord"
conditioning_varb = "Distance"
num_closest = 3

closest_stim = stim_conditionals(stim_data, conditioning_varb, varb, num_closest)
closest_random_stim = stim_conditionals(random_stim_data, conditioning_varb, varb, num_closest)

fig, attended, whole_dist, rank = stim_analyzer(stim_data, varb, hittypes, 0, alt_yellow, [closest_random_stim])
fig, attended, whole_dist, rank = stim_analyzer(stim_data, varb, hittypes, 0, alt_yellow, [closest_stim])
```

```python
varb = "Az Coord"
conditioning_varb = "Distance"
num_closest = 1
hittypes = [1,2,3]
closest_stim = stim_conditionals(stim_data, conditioning_varb, varb, num_closest)
closest_random_stim = stim_conditionals(random_stim_data, conditioning_varb, varb, num_closest)
fig, attended, whole_dist, rank = stim_analyzer(stim_data, varb, hittypes, 0, az_blue, [closest_random_stim])
fig, attended, whole_dist, rank = stim_analyzer(stim_data, varb, hittypes, 0, az_blue, [closest_stim])
```
